# PyTrebuchet: Optimize a trebuchet design

This notebook finds the optimum design of a trebuchet, given some constraints and bounds on the design variables

## 1. Setup

```python
import sys

sys.path.insert(0, "../src")
from pytrebuchet import DesignOptimizer
import numpy as np

# Create the design optimizer with specified parameters and constraints
optimizer = DesignOptimizer(
    # Define fixed parameters and design variables (initial guesses, bounds)
    length_arm=1.75 + 6.792,  # fixed total arm length
    mass_arm=10,
    mass_projectile=4.0,
    diameter_projectile=0.35,
    fraction_projectile_arm=(0.795, (0.5, 0.9)),
    mass_counterweight=98.09,
    length_sling_projectile=(6.833, (1.0, 10.0)),
    length_sling_weight=2.0,
    height_pivot=5.0,
    release_angle=(45 * np.pi / 180, (25 * np.pi / 180, 60 * np.pi / 180)),
    # Define constraints
    constrain_sling_tension=True,  # Ensure sling remains in tension
)
```

## 2. Run the optimizer and print the results.

```python
results = optimizer.optimize()
print("Optimization Results:")
print(f"Optimized distance: {-results.fun:.2f} meters")
print("Optimized parameters:")
print(f"Projectile arm length: {(1.75 + 6.792) * results.x[0]:.2f} meters")
print(f"Weight arm length: {(1.75 + 6.792) * (1 - results.x[0]):.2f} meters")
print(f"Projectile sling length: {results.x[1]:.2f} meters")
print(f"Release angle: {results.x[2] * 180 / np.pi:.2f} degrees")
```

## 3. Compare to the baseline trebuchet.

```python
print(f"Baseline distance: {optimizer.get_baseline_distance():.2f} meters")
print(
    f"Ratio of optimized to baseline distance: {-results.fun / optimizer.get_baseline_distance():.2f}"
)
```
